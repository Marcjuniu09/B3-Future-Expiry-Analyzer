```python
import requests
import pandas as pd

# Define a chave de API da Alpha Vantage
api_key = "99T6GTGZDMJC0L19"
# Define o código da ação
codigo_da_acao = "PETR4.SA"

# Define a função para obter os dados da API
def obter_dados_alpha_vantage(api_key, codigo_da_acao):
    url = f"https://www.alphavantage.co/query?function=TIME_SERIES_DAILY_ADJUSTED&symbol={codigo_da_acao}&apikey={api_key}&outputsize=full"
    response = requests.get(url)
    dados = response.json()['Time Series (Daily)']
    df = pd.DataFrame(columns=["Data", "Abertura", "Alta", "Baixa", "Fechamento", "Volume"])
    for dia, valores in dados.items():
        data = pd.to_datetime(dia)
        abertura = float(valores['1. open'])
        alta = float(valores['2. high'])
        baixa = float(valores['3. low'])
        fechamento = float(valores['4. close'])
        volume = int(valores['6. volume'])
        df.loc[data] = [data, abertura, alta, baixa, fechamento, volume]
    df.sort_index(inplace=True)

    # Adiciona uma coluna para o SMA de 20 dias
    periodo = 20
    sma = df['Fechamento'].rolling(window=20).mean()
    df[f'SMA{periodo}'] = sma

    return df

# Chama a função para obter os dados
petr4 = obter_dados_alpha_vantage(api_key, codigo_da_acao)
#salvando em um arquivo csv
petr4.to_csv('exemplo.csv', index=False)
# Exibe os dados obtidos
display(petr4)
```
